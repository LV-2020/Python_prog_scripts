```python
def divisible(n):
    result = []
    for i in range(1,n):
        if i %6==0:
            result.append(i)
    return result

divisible(20)
```

```python
def sumbythree(n):
    sum = 0
    for i in range(1,n):
        if i%3==0:
            sum=sum+i
    return sum

sumbythree(20)
            
```

```python


def printprimes(n):
    primes = []
    count = 0
    for n in range(2,n):
        is_prime = True
        for i in range(2,n):
            if n%i==0:
                is_prime = False
                break
        if is_prime:
            count = count + 1
            primes.append(n)
        
    return primes, count

printprimes(20)
```

```python
# Find largest prime number less than n
def largest_prime(n):
    primes = []
    for n in range(n-1,1,-1):
        is_prime = True
        for i in range(n-1,1,-1):
           if n%i==0:
               is_prime = False
    
           if is_prime:
               primes.append(n)
               break
               
        return primes

largest_prime(32)
```

```python
def is_palindrome(n):
    original = n
    reverse = 0

    while n>0:
        digit = n % 10
        reverse = reverse *  10 + digit
        n = n // 10

    return original == reverse

print(is_palindrome(121))
                    
```

```python
def target9(nums,target):
    for i in range(len(nums)):
        for j in range(i+1,len(nums)):
            if nums[i] + nums[j] == target:
                return [i,j]

target9([1,2,3,4,5],9)
```

```python

def  two_sum(nums,target):
    seen = {}

    for i in range(len(nums)):
        complement = target - nums[i]

        if complement in seen:
            return [seen[complement],i]
        
        seen[nums[i]] = i


two_sum([1,2,3,4,5],9)
```

```python
### Find the largest of numbers
def findlargest(numbers):
    largest = 0
    for i in range(0,len(numbers)):
        for j in range(1,len(numbers)):
            if numbers[i] > numbers[j]:
                largest = numbers[i]
            else:
                largest = numbers[j]
    print(largest)

findlargest([1,2,6,4,5])
```

```python
## Remove duplcates
def remove_duplicates(list):

    result = []
    seen = set()

    for i in list:
        if i not in seen:
            result.append(i)
            seen.add(i)
    print(result)

remove_duplicates([1,2,2,3,3,4,5])
```
