```python
### How to identify odd numbers between one to ten in Python
def print_odds():
    for i in range(1,20):
        if i % 2 != 0:
            print(i)
print_odds()
```

```python
# Print all even numbers between 1 to 20
#1) Numbers should be between 1 and 10
#2) for each number between 1 and 10
#3) if i%2==0
#4) then it is an even number 
#5) print the number
for i in range(1,20):
    if i%2==0:
        print(i)
```

```python
def divisible(n):
    result = []
    for i in range(1,n):
        if i %6==0:
            result.append(i)
    return result

divisible(20)
```

```python
def sumbythree(n):
    sum = 0
    for i in range(1,n):
        if i%3==0:
            sum=sum+i
    return sum

sumbythree(20)
            
```
