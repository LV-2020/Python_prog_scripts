```python
# List containing duplicates
# lets say there is a list a = [1,2,3,4,1]
# How do we identify duplicates
# First go through the list 
# Iterate over a for loop
#If a[i]=a[j] then thats the duplicate and print that

def duplicates(nums):
    is_duplicate = False
    for i in range(len(nums)):
        for j in range(len(nums)):
            if nums[i]==nums[j]:
                return True
                print("Duplicate found")


```

```python
#find the sum of elements in a list

def sum_list(nums):
    sum=0
    for i in range(0,len(nums)):
        sum = sum + nums[i]
    
    print(sum)
```
