```python
### Use a dictionary to count the frequency of elements in a list
numbers = [1,2,2,3,3,4,4,5]
frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1

print(frequency)
```

```python
### How to find the frequneyc of elements in a list
def freq(s):
    d = {}
    for ch in s:
        d[ch] = d.get(ch,0) + 1
    return d

freq([1,22,33,2,4])
```

```python
def max_track(lst):
    m = float('-inf')
    for x in lst:
        if x > m:
            m = x
    return m

max_track([1,2,2,3,3,4])
```

```python
employees = [
                {"name":"Vas","salary":3000},
                {"name": "Edi","salary":20000},
                {"name":"Sas","salary":45}
]

def top_two(employees):
    sorted_list = sorted(employees,key = lambda x:x['salary'],reverse = True)
    return sorted_list[-2]

top_two(employees)
```

```python
def calculate(cart):
    total_cost = 0
    for item in cart:
        total_cost += item['price'] * item['quantity']
    return total_cost

cart = [
        {"name":"apple","price":0.5,"quantity":4},
        {"name":"banana","price":0.3,"quantity":10},
        {"name":"orange","price":0.7,"quantity":15} 
]

total_cost = calculate(cart)
print(total_cost)
```

```python
### Create a function to check if a string is a palindrome
def is_palindrome(str):
    str = str.lower().replace(" ","")
    return str == str[::-1]

print(is_palindrome("aba"))
```

```python
## Calculate the factorial of a given number using recursion
# function that accepts a number as a parameter
# factorial of 4 = 4*3*2*1
# Iteate through the numbers using for loop
# prod = prod * factorial (i-1)
# finally print prod

def factorial(n):

    if n<0 : 
        print("Out of bound error")
        return None
    
    if n==0 or n==1:
        return 1
    

    return  n * factorial(n-1)


factorial(3)
```

```python
### Write s function to return the ascending order of integers
def ascendingorder(a,b,c):
    if a<b and c<b:
        if a>c :
            print(c,a,b)
        else:
            print(a,c,b)
    elif a<c and b<c:
        if a>b : 
            print(b,a,c)
        else:
            print(a,b,c)
    elif c<a and b<a:
        if c>b:
            print(b,c,a)
        else:
            print(c,b,a)

    else:
        print(a,b,c)

ascendingorder(3,2,4)
```

```python
def generate_triangle(n):
    ch = "*"
    pat = []

    for i in range(1,n+1):
        pat.append(" " * (n-i)+ ch * i)
    return pat

generate_triangle(4)
```

```python
def largest_numb(numbers):
    largest = numbers[0]

    for num in numbers:
        if num > largest:
            largest = num
    
    print(largest)

largest_numb([1,3,4,7,5])
```
