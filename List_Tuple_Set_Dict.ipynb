```python
### How to find the frequneyc of elements in a list
def freq(s):
    d = {}
    for ch in s:
        d[ch] = d.get(ch,0) + 1
    return d

freq([1,22,33,2,4])
```

```python
## Calculate the factorial of a given number using recursion
# function that accepts a number as a parameter
# factorial of 4 = 4*3*2*1
# Iteate through the numbers using for loop
# prod = prod * factorial (i-1)
# finally print prod

def factorial(n):

    if n<0 : 
        print("Out of bound error")
        return None
    
    if n==0 or n==1:
        return 1
    

    return  n * factorial(n-1)


factorial(3)
```

```python
### Write s function to return the ascending order of integers
def ascendingorder(a,b,c):
    if a<b and c<b:
        if a>c :
            print(c,a,b)
        else:
            print(a,c,b)
    elif a<c and b<c:
        if a>b : 
            print(b,a,c)
        else:
            print(a,b,c)
    elif c<a and b<a:
        if c>b:
            print(b,c,a)
        else:
            print(c,b,a)

    else:
        print(a,b,c)

ascendingorder(3,2,4)
```

```python
### Rotating the list
def rotate(lst,k):
    n = len(lst)
    result = []

    if k>n:
        return lst
    else:
        k=k%n

    for i in range(n-k,n):
        result.append(lst[i])
    for i in range(0,n-k):
        result.append(lst[i])

    return result

rotate([1,2,3,4],2)
```

```python
# Count the frequency of elements usingh a dictionary

def count_dict(lst):
    result = {}
    value = 1
    
    for i in range(len(lst)):
        if lst[i] in result:
           result[lst[i]] = value + 1
        if lst[i] not in result:
            result[lst[i]] = value
            
    return result

count_dict([1,3,3,4])
```
