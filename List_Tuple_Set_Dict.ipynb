```python
### Use a dictionary to count the frequency of elements in a list
numbers = [1,2,2,3,3,4,4,5]
frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1

print(frequency)
```

```python
### How to find the frequneyc of elements in a list
def freq(s):
    d = {}
    for ch in s:
        d[ch] = d.get(ch,0) + 1
    return d

freq([1,22,33,2,4])
```

```python
def max_track(lst):
    m = float('-inf')
    for x in lst:
        if x > m:
            m = x
    return m

max_track([1,2,2,3,3,4])
```

```python
def evens_squared(lst):
    return [x*x for x in lst if x%2==0]

evens_squared([1,2,3,4])
```

```python
employees = [
                {"name":"Vas","salary":3000},
                {"name": "Edi","salary":20000},
                {"name":"Sas","salary":45}
]

def top_two(employees):
    sorted_list = sorted(employees,key = lambda x:x['salary'],reverse = True)
    return sorted_list[-2]

top_two(employees)
```
